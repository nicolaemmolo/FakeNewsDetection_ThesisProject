```python
TOPIC_ORDER = ['politics', 'general', 'covid', 'syria', 'islam', 'notredame', 'gossip'] # Topic order for plots
DATE_ORDER = ['2011-2015', '2016', '2017', '2019', '2020'] # Date order for plots
MODEL_ORDER = ['BERT', 'DeBERTa', 'RoBERTa', 'CNN', 'BiLSTM', 'Linear', 'NB', 'PA'] # Model order for plots
TECHNIQUE_ORDER = [
    'Finetuning',
    'Replay (n=50)',
    'Replay (n=100)',
    'Replay (n=200)',
    'LwF (α=03)',
    'LwF (α=05)',
    'EWC (λ=300)',
    'EWC (λ=500)',
    'EWC (λ=1000)',
    'Hybrid (α=03, n=50)',
    'Hybrid (α=03, n=100)',
    'Hybrid (α=03, n=200)',
    'Hybrid (α=05, n=50)',
    'Hybrid (α=05, n=100)',
    'Hybrid (α=05, n=200)'
]
```

```python
# 1. Carica tutti i dati per il modello scelto
target_model = 'RoBERTa'
type = 'date' # 'topic' or 'date'
type_order = TOPIC_ORDER if type == 'topic' else DATE_ORDER
all_tech_data = load_data_for_model_comparison(target_model, type=type)

if target_model not in ['Linear', 'NB', 'PA']:
    # 2. Get data for Replay (+ Finetuning)
    data_replay = {k: v for k, v in all_tech_data.items() if 'Replay' in k or k == 'Finetuning'}

    # 3. Get data for Distillation (+ Finetuning)
    data_distillation = {k: v for k, v in all_tech_data.items() if 'LwF' in k or k == 'Finetuning'}

    # 4. Get data for EWC (+ Finetuning)
    data_ewc = {k: v for k, v in all_tech_data.items() if 'EWC' in k or k == 'Finetuning'}

    # 5. Get data for Hybrid (+ Finetuning)
    data_hybrid = {k: v for k, v in all_tech_data.items() if 'Hybrid' in k or k == 'Finetuning'}
```

```python
# ALL TECHNIQUES
plot_full_comparison_model(target_model, all_tech_data, type_order, type=type)

if target_model not in ['Linear', 'NB', 'PA']:
    # REPLAY
    plot_full_comparison_model(target_model, data_replay, type_order, type=type)

    # DISTILLATION
    plot_full_comparison_model(target_model, data_distillation, type_order, type=type)

    # EWC
    plot_full_comparison_model(target_model, data_ewc, type_order, type=type)

    # HYBRID
    plot_full_comparison_model(target_model, data_hybrid, type_order, type=type)
```
